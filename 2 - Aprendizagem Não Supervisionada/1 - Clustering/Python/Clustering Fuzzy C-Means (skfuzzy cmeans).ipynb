```python
from sklearn import datasets
import numpy as np
from sklearn.metrics import confusion_matrix
import skfuzzy
```

```python
iris = datasets.load_iris()
```

```python
iris.data.shape
```

```python
iris.target
```

```python
iris.target_names
```

```python
unicos, quantidade = np.unique(iris.target, return_counts = True)
unicos, quantidade
```

```python
r = skfuzzy.cmeans(data = iris.data.T, c=3 , m=2, error=0.005 , maxiter=1000, init=None)
# T -> Transpose
# c -> Número de clusters
# m -> Membership (default 2)
# error (default 0.005)
# maxiter -> Máximo de iterações
# init (default None)
r
# Retorna 
# cntr - Centro de cada cluster
# u - Resultados
# u0 - Ponto de partida
# d - Distâncias euclidianas
# jm - Histórico da função
# p - Número de iterações realizadas
# fpc - Coeficiente de particionamento difuso
```

```python
previsoes_porcentagem = r[1] # u - Resultados
previsoes_porcentagem
```

```python
previsoes_porcentagem[0][0] 
#[Primeiro Cluster][Primeiro registro] - Probabilidade de fazer parte desse grupo
```

```python
previsoes_porcentagem[1][0]
#[Segunda Cluster][Primeiro registro] - Probabilidade de fazer parte desse grupo
```

```python
previsoes_porcentagem[2][0]
#[Terceiro Cluster][Primeiro registro] - Probabilidade de fazer parte desse grupo
```

```python
previsoes = previsoes_porcentagem.argmax(axis=0)
previsoes
```

```python
resultados = confusion_matrix(iris.target, previsoes)
resultados
```
