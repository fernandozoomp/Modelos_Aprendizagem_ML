### Clustering K-Means

```r
dim(iris)
```

```r
head(iris)
```

```r
summary(iris)
```

```r
cluster = kmeans(iris[1:4], center=3)
cluster
```

```r
table(iris$Species, cluster$cluster)
```

```r
options(repr.plot.width=8, repr.plot.height=8, repr.plot.res=300)
plot(iris[,1:4], col=cluster$cluster)
```
