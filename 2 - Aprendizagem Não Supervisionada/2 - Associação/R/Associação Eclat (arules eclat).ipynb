### Associação Eclat

```r
library(arules)
```

```r
transacoes = read.transactions(file.choose(), format="basket", sep=" ") # transacoes2.txt
transacoes
```

```r
image(transacoes)
```

```r
regras = eclat(transacoes, parameter=list(supp=0.1, maxlen=15))
```

```r
inspect(regras)
```

```r
library(arulesViz)
```

```r
options(repr.plot.width=6, repr.plot.height=6, repr.plot.res=300)
plot(regras, method="graph", control=list(type="items"))
```
